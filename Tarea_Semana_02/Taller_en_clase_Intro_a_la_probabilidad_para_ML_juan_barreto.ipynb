## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números: 
1. Grafique los números aleatorios, 
2. Ordene de menor a mayor y grafique,
4. Encuentre los valores teoricos de media y varianza, y comparelos con los estimados.

```python
#Para todas las simulaciones:
N=100000
m=100
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# bernoulli con cierto p entre 0 y 1
p=.7
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
r = bernoulli.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
```

```python
print(p)
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
```

```python
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
b = binom.rvs(n, p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.binom.html#scipy.stats.binom
plt.plot(b)
plt.title(str(N)+" puntos distribuidos binomial con n=" + str(n) + " con p="+str(p))
plt.show()
```

```python
print(n, p)
b.sort()
plt.plot(b)
plt.title(str(N)+" puntos distribuidos binomial ordenados con n=" + str(n) +" y p="+str(p))
plt.show()
```

```python
plt.hist(b, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomiales con n= "+str(n) + " y con p="+str(p))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de b")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=b.mean()
varest=b.var()
print("Los valores teoricos de la media y varianza para la binomial con n=", n, "con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la binomial con n=", n, " con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

## Segundo punto Distribuciones contínuas.


```python
#Para todas las simulaciones:
N=1000000
m=100
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
from scipy.stats import norm
mean, var, skew, kurt = norm.stats(moments='mvsk',loc=u,scale=std)
nm = norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.norm.html#scipy.stats.norm
plt.plot(nm)
plt.title(str(N)+" puntos distribuidos normalmente con media="+str(u)+" y std="+str(std))
plt.show()
nm.sort()
plt.plot(nm)
plt.title(str(N)+"puntos distribuidos normalmente ordenados con media="+str(u)+" y std="+str(std))
plt.show()
plt.hist(nm, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normalmente con u="+str(u)+ " y std= "+str(std))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo nm:")
print(type(nm))
print("Vamos a hallar algunas estimaciones de estadisticas de nm")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=nm.mean()
varest=nm.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Uniforme entre 0 y 255
a=0
b=255
from scipy.stats import uniform
mean, var, skew, kurt = uniform.stats(moments='mvsk',loc=a,scale=b)
um = uniform.rvs (loc=a,scale=b, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform
plt.plot(um)
plt.title(str(N)+" puntos distribuidos uniformente con media="+str(a)+" y std="+str(b))
plt.show()
um.sort()
plt.plot(um)
plt.title(str(N)+" puntos distribuidos uniformente ordenados con u="+str(a)+ " y std= "+str(b))
plt.show()
plt.hist(um, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos uniformente con u="+str(a)+ " y std= "+str(b))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=um.mean()
varest=um.var()
print("Los valores teoricos de la media y varianza para la uniforme son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Exponencial con Lamnda L
L=0.3
from scipy.stats import expon
mean, var, skew, kurt = expon.stats(moments='mvsk',scale=1/L)
e = expon.rvs(scale=1/L, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.expon.html#scipy.stats.expon
plt.plot(e)
plt.title(str(N)+" puntos distribuidos exponencialmente con L="+str(L))
plt.show()
e.sort()
plt.plot(e)
plt.title(str(N)+" puntos distribuidos exponencialmente ordenados con L="+str(L))
plt.show()
plt.hist(e, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos exponencialmente con L="+str(L))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de e")
avg=e.mean()
varest=e.var()
print("Los valores teoricos de la media y varianza para la exponencial son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
#Chi cuadrada con K grados de libertad
K=3.0
from scipy.stats import chi2
mean, var, skew, kurt = chi2.stats(K, moments='mvsk')
c = chi2.rvs(K, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.chi2.html#scipy.stats.chi2
plt.plot(c)
plt.title(str(N)+" puntos distribuidos chi cuadrado con K="+str(K))
plt.show()
c.sort()
plt.plot(c)
plt.title(str(N)+" puntos distribuidos chi cuadrado ordenados con K="+str(K))
plt.show()
plt.hist(c, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos chi cuadrado con K="+str(K))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de c")
avg=c.mean()
varest=c.var()
print("Los valores teoricos de la media y varianza para la chi cuadrado son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
#Gamma con parametros k y theta t
k,t=2.54,0.43
from scipy.stats import gamma
mean, var, skew, kurt = gamma.stats(a=k,scale=t, moments='mvsk')
gmm = gamma.rvs(a=k,scale=t, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.chi2.html#scipy.stats.chi2
plt.plot(gmm)
plt.title(str(N)+" puntos distribuidos en Gamma con K="+str(K))
plt.show()
gmm.sort()
plt.plot(gmm)
plt.title(str(N)+" puntos distribuidos en Gamma ordenados con K="+str(K))
plt.show()
plt.hist(gmm, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos en Gamma con K="+str(K))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de c")
avg=gmm.mean()
varest=gmm.var()
print("Los valores teoricos de la media y varianza para la Gamma son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```
